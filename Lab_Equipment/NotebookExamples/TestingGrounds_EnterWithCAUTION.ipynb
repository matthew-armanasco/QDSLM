```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Example multi-peak data
x = np.linspace(0, 10, 1000)
y = (np.sin(3 * x) ** 2) + 0.3 * np.sin(10 * x) + 0.2 * np.random.randn(len(x))

# Find peaks
peaks, properties = find_peaks(y, height=0.1, distance=100)

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Signal')
plt.plot(x[peaks], y[peaks], "x", label='Peaks')
plt.title('Detected Peaks in Multi-Peak Distribution')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# Optional: print peak info
for i, peak in enumerate(peaks):
    print(f"Peak {i+1}: x = {x[peak]:.3f}, y = {y[peak]:.3f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Generate sample data: a combination of sine waves with added noise
x = np.linspace(0, 10, 1000)
y = np.sin(x) + np.sin(3 * x) + 0.3 * np.random.normal(size=x.size)

# Find peaks in the data; adjust parameters such as 'height' and 'distance' as needed
peaks, properties = find_peaks(y, height=0)  # 'height=0' ensures we capture peaks above 0

# Print the indices and the corresponding peak values
print("Peak indices:", peaks)
print("Peak values:", y[peaks])

# Plot the original signal and mark the detected peaks
plt.figure(figsize=(10, 5))
plt.plot(x, y, label="Signal")
plt.plot(x[peaks], y[peaks], "x", label="Detected Peaks", markersize=10)
plt.title("Multi-Peak Distribution with Detected Peaks")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
import cv2
import numpy as np
import tkinter as tk
import threading
from screeninfo import get_monitors

def opencv_display_on_monitor(monitor_index=0):
    # Retrieve information about all connected monitors
    monitors = get_monitors()
    if monitor_index >= len(monitors):
        print(f"Monitor index {monitor_index} out of range. Using primary monitor instead.")
        monitor = monitors[0]
    else:
        monitor = monitors[monitor_index]

    # Get the monitor's position and dimensions
    monitor_x = monitor.x
    monitor_y = monitor.y
    monitor_width = monitor.width
    monitor_height = monitor.height
    print(f"Using monitor {monitor_index}: x={monitor_x}, y={monitor_y}, width={monitor_width}, height={monitor_height}")
    return monitor_x,monitor_y,monitor_height, monitor_width
def opencv_display_v2(monitor_x,monitor_y,monitor_height, monitor_width):
     # Create a blank image sized to the monitor's resolution
    img = np.zeros((monitor_height, monitor_width, 3), dtype=np.uint8)
    # Set the right half of the image to red (BGR format: red is [0,0,255])
    img[:, monitor_width // 2:, :] = [0, 0, 255]

    # Create a window that we can position manually
    cv2.namedWindow("Display", cv2.WINDOW_NORMAL)
    # Set window to full screen mode if desired
    cv2.setWindowProperty("Display", cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)
    # cv2.resizeWindow("Display", monitor_width, monitor_height)
    # Move the window to the desired monitor's position
    # cv2.moveWindow("Display", monitor_x, monitor_y)

    # Move the window to the monitor's position
    cv2.moveWindow("Display", monitor_x, monitor_y)

    # Display loop
    while True:
        cv2.imshow("Display", img)
        # Break loop when the ESC key is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break

    cv2.destroyAllWindows()
    cv2.waitKey(1)
    cv2.destroyAllWindows()
    
def get_screen_resolution():
    root = tk.Tk()
    root.withdraw()  # Hide the main tkinter window
    width = root.winfo_screenwidth()
    height = root.winfo_screenheight()
    root.destroy()
    return width, height

def opencv_display():
    width, height = get_screen_resolution()
    # Create a blank black image
    img = np.zeros((height, width, 3), dtype=np.uint8)
    # Set the right half of the image to red
    img[:, width//2:, :] = [0, 0, 255]

    # Create a full-screen window
    cv2.namedWindow("Display", cv2.WINDOW_NORMAL)
    cv2.setWindowProperty("Display", cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

    while True:
        cv2.imshow("Display", img)
        # Wait 1 ms for a key event; exit if 'ESC' is pressed
        if cv2.waitKey(1) == 27:  # 27 is the ASCII code for ESC
            break
    cv2.destroyAllWindows()
    cv2.waitKey(1)
    cv2.destroyAllWindows()
    
    
```

```python
import cv2
import numpy as np
import time
from AppKit import NSWorkspace, NSScreen
from Quartz import CGWindowListCopyWindowInfo, kCGWindowListOptionOnScreenOnly, kCGNullWindowID
import subprocess

def get_screens():
    return NSScreen.screens()

def get_opencv_window_id(window_name):
    window_list = CGWindowListCopyWindowInfo(kCGWindowListOptionOnScreenOnly, kCGNullWindowID)
    for window in window_list:
        if window_name in window.get('kCGWindowName', ''):
            return window['kCGWindowNumber']
    return None

def make_window_fullscreen_on_display(window_name, display_index=0):
    screens = get_screens()
    if display_index >= len(screens):
        raise ValueError("Display index out of range.")
    
    screen = screens[display_index]
    frame = screen.frame()
    x, y = int(frame.origin.x), int(frame.origin.y)
    width, height = int(frame.size.width), int(frame.size.height)

    # Use AppleScript to move and resize the OpenCV window
    script = f'''
    tell application "System Events"
        set frontmostProcess to first process whose frontmost is true
        tell frontmostProcess
            set theWindow to first window whose name contains "{window_name}"
            set position of theWindow to {{{x}, {y}}}
            set size of theWindow to {{{width}, {height}}}
        end tell
    end tell
    '''
    subprocess.call(["osascript", "-e", script])
```

```python
# Set up the OpenCV window
window_name = "FullscreenImage"
cv2.namedWindow(window_name, cv2.WINDOW_NORMAL)

# Show dummy frame to register window
cv2.imshow(window_name, np.zeros((100, 100, 3), dtype=np.uint8))
cv2.waitKey(1)
time.sleep(1)  # allow window to appear

# Move window to secondary display (e.g. display index 1)
make_window_fullscreen_on_display(window_name, display_index=0)

# Start image loop
while True:
    img = np.random.randint(0, 256, (1080, 1920, 3), dtype=np.uint8)
    cv2.imshow(window_name, img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
cv2.waitKey(3)
cv2.destroyAllWindows()
```
